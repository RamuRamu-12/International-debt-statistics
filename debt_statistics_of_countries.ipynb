```python
fig = px.scatter(data, x="indicator_code", y="debt", color="country_name")
fig.show()
```

```python
total_debt_by_country = data.groupby('country_name')['debt'].sum().reset_index()
```

```python
total_debt_by_country
```

```python
fig = px.bar(total_debt_by_country, x='country_name', y='debt', title='Total Debt by Country')
fig.update_xaxes(categoryorder="total descending")  # Sort bars by total debt values

# Show the plot
fig.show()
```

```python
max_debt_row = total_debt_by_country.loc[total_debt_by_country['debt'].idxmax()]
```

```python
max_debt_row
```

```python
avg_debt_by_country = data.groupby('country_name')['debt'].mean().reset_index()
```

```python
avg_debt_by_country.head(5)
```

```python
fig = px.bar(avg_debt_by_country, x='country_name', y='debt', title='Avg Debt by Country')
fig.update_xaxes(categoryorder="total descending")  # Sort bars by avg debt values

# Show the plot
fig.show()
```

```python
# Finding the highest amount of principal repayments
max_principal_repayment = data['debt'].max()

# Finding the row with the highest principal repayment and extract the corresponding indicator name and country name
max_principal_repayment_row = data[data['debt'] == max_principal_repayment]
max_principal_repayment_country = max_principal_repayment_row['country_name'].values[0]
max_principal_repayment_indicator = max_principal_repayment_row['indicator_name'].values[0]

# Displaying the highest amount of principal repayments and the corresponding indicator and country
print("Highest Principal Repayment:", max_principal_repayment)
print("Country with Highest Principal Repayment:", max_principal_repayment_country)
print("Indicator Name for Highest Principal Repayment:", max_principal_repayment_indicator)
```

```python
# Find the most common debt indicator
most_common_debt_indicator = data['indicator_name'].values[0]

# Display the most common debt indicator
print("Most Common Debt Indicator:", most_common_debt_indicator)
```
